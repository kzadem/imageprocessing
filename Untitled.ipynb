```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
image = cv2.imread("manzara.jpg")
```

```python
image = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)
```

```python
pixel_values = image.reshape((-1, 3))

pixel_values = np.float32(pixel_values)
```

```python
print(pixel_values.shape)
```

```python
criteria = (cv2.TERM_CRITERIA_EPS + cv2.TERM_CRITERIA_MAX_ITER, 100, 0.2)
```

```python
k = 3
_, labels, (centers) = cv2.kmeans(pixel_values, k, None, criteria, 10, cv2.KMEANS_RANDOM_CENTERS)
```

```python
centers = np.uint8(centers)

labels = labels.flatten()
```

```python
segmented_image = centers[labels.flatten()]
```

```python
 = segmented_image.reshape(image.shape)
plt.imshow(segmented_image)
plt.show()
```

```python
segmentationimageone=cv2.divide(image,segmented_image, scale=256.0)
```

```python
cv2.imwrite("segmentationimageone.png", segmentationimageone)
```

```python
segmentationimageone = segmentationimageone.reshape((-1, 3))
cluster = 2
segmentationimageone[labels == cluster] = [0, 0, 0]
segmentationimageone = segmentationimageone.reshape(image.shape)
plt.imshow(segmentationimageone)
plt.show()
```
